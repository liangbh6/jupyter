## func

```r
func <- function(a){
    strings = paste(a[-1], collapse='')  # convert to string to count
    b = as.data.frame(table(strsplit(strings,split='')))  # count
    res = str_count(a[-1], as.character(b$Var1[which.min(b$Freq)]))
return(res)}
```

```r
results = apply(genotype,2,func)
dim(results)
```

```r
phenotype = read.table('./data/phenotype.txt')
```

```r
set.seed(1)
write.table(results,"./data/genotype_01encoing.csv")
```

```r
library(randomForest)
mydata <- cbind(phenotype, results)
y = mydata[,1]
x = mydata[,-1]
```

```r
rf <-randomForest(x=x,y=factor(y),mtry=97,ntree=200,importance=TRUE)
```
