# random forest

```r
library(randomForest)
genotype_<-genotype[2:1001,]
phenotype = read.table('./data/phenotype.txt')
mydata<-cbind(phenotype,genotype_)
y<-mydata[,1]
x<-mydata[,-1]
set.seed(1)
```

## 此处选择分类树,mtry=sqrt(9445),即mtry取97，选择200颗树


```r
rf <-randomForest(x=x,y=factor(y),mtry=97,ntree=200,importance=TRUE)
```

```r
im<-matrix()
im<-importance(rf)
order4<-im[order(-im[,4]),] ##按照第4列降序
orderall<-order4[1:10,]     
orderall<-write.table(orderall,"D:/orderall.txt")
```
